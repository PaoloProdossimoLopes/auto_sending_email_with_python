```
#Passos importantes caso de erro na execução:
# PASSO 1-> Habilitar o acessor ao email: https://accounts.google.com/signin/continue?sarp=1&scc=1&plt=AKgnsbt\n5.7.14
# PASSO 2-> permitir o google: https://support.google.com/mail/answer/78754

#Importando a biblioteca para manusear as tabelas--------------------------------------------------------------------------------------------------------------------------------
import pandas as pd

#Recuperando tabela--------------------------------------------------------------------------------------------------------------------------------
data_frame = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/SLAutoReports/Vendas 2021.xlsx", sheet_name="Vendas")
tabela_padrao = data_frame

#Tratando dados--------------------------------------------------------------------------------------------------------------------------------

replace_NaN = "0.0"
replace_media = "Medias: "

#exclusao de colunas
tabela_padrao = tabela_padrao.drop("Unnamed: 0", axis= 1)
tabela_padrao = tabela_padrao.drop("Unnamed: 6", axis= 1)
tabela_padrao = tabela_padrao.drop("Unnamed: 7", axis= 1)
tabela_padrao = tabela_padrao.drop("Unnamed: 8", axis= 1)
tabela_padrao = tabela_padrao.drop("Unnamed: 9", axis= 1)
tabela_padrao = tabela_padrao.drop("Unnamed: 10", axis= 1)
tabela_padrao = tabela_padrao.drop("Unnamed: 11", axis= 1)

# Nomenando Colunas
tabela_padrao.columns = ['Mes', 'Dias', 'Vendas','Vendas diarias','Vendas diarias por trimestre']

#excluido linhas extras
for linhas in tabela_padrao.index:
  if linhas > 14: 
    tabela_padrao = tabela_padrao.drop(linhas, axis = 0)

#excluindo dados inicias
tabela_padrao = tabela_padrao.drop(0, axis= 0)
tabela_padrao = tabela_padrao.drop(1, axis= 0)

#formatando valores da tabela
tabela_padrao['Vendas diarias por trimestre'].fillna(replace_NaN, inplace=True)
tabela_padrao['Mes'].fillna(replace_NaN, inplace=True)
tabela_padrao['Vendas'].fillna(replace_NaN, inplace=True)
tabela_padrao['Vendas diarias'].fillna(replace_NaN, inplace=True)
tabela_padrao['Dias'].fillna(replace_NaN, inplace=True)

#Mudando a variavel
tabela_tratada = tabela_padrao

display(tabela_tratada)

#função de enviar email (Snippet)--------------------------------------------------------------------------------------------------------------------------------
import smtplib
import email.message
import getpass as gp

def enviar_email(resumo_loja, referente, array_emials, senha_do_email, assunto, email_de_envio, usuario):

  server = smtplib.SMTP('smtp.gmail.com:587')  
  email_content = f'''
  <p>Segue a baixo a o Relatorio referente ha {referente},</p>
  {resumo_loja.to_html()}
  <p>Qualquer duvida entrar em contato,</p>
  <p>Atenciosamente,</p>
  <p>{usuario}</p>
  '''

  titulo_do_email = assunto          
  
  msg = email.message.Message()
  msg['Subject'] = f"{titulo_do_email} - {referente}"
  
  msg['From'] =  email_de_envio       
  msg['To'] = array_emials            
  
  password = senha_do_email

  msg.add_header('Content-Type', 'text/html')
  msg.set_payload(email_content)
  
  s = smtplib.SMTP('smtp.gmail.com: 587')
  s.starttls()

  # Login Credentials for sending the mail
  s.login(msg['From'], password)
  s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))

  print(f"email para {msg['To']} enviado com sucesso!")


#Lista de emails que irão receber o relatorio--------------------------------------------------------------------------------------------------------------------------------
array_emails = ["prols.developer+1@gmail.com","prols.developer+2@gmail.com","prols.developer+3@gmail.com","prols.developer+4@gmail.com","prols.developer+5@gmail.com"]

#Credenciais--------------------------------------------------------------------------------------------------------------------------------------------------------------------

print("Faça o Login:")

usuario = input("Quem está enviando?: ")

email_entrada = input("Email: ")

print("Senha: ")
senha_entrada = gp.getpass()

assunto = input("Assunto: ")
referente = input("referente ha: ")

# Mensagem de Status-----------------------------------------------------------------------------------------------------------------------------------------------------------
print("Enviando emails...")

for itens in array_emails:
  enviar_email(tabela_tratada, referente, itens, senha_entrada, assunto, email_entrada, usuario)

print("Processo finalizado com sucesso!")
```
